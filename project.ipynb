```python
import numpy as np

def take_input():
  choice = str.upper(input('Podaj figure (K, P, N): '))
  assert choice in ['K','P','N'], 'WRONG INPUT, SHOULD BE ONE OF: K, P, N'

  return choice

def peek(A):
  return A[len(A) - 1]

def manage_registry(registry, c, h):
  if response[h] == c:
    registry.append(peek(registry)+1)
  elif h == c:
    registry.append(peek(registry))
  else:
    registry.append(peek(registry)-1)


'''
                    nastepny
                      K   P   N
                    K[1/3,1/3,1/3]
poprzedni           P[1/3,1/3,1/3]
                    N[1/3,1/3,1/3]

'''              
prop_matrix = np.array([[1/3,1/3,1/3],
                         [1/3,1/3,1/3],
                         [1/3,1/3,1/3]])

prop_dict = {'K' : np.array([0,0,0]),
             'P' : np.array([0,0,0]),
             'N' : np.array([0,0,0])}
#      key is beaten by value 
response = {'K' : 'P', #P beats K
            'P' : 'N', #N beats P 
            'N' : 'K'} #K beats N

registry = [0]

human_choice = 'K'#take_input()
previous = human_choice

comp_choice = 'P'#np.random.choice(['K','P','N'], p=[1/3,1/3,1/3])

'''
jeżeli wiemy, że w zerowym przypadku wypadło K
a w następnym człowiek wybrał K 
'''






for i in range(250):
  
  print('\n\n\n')
  print(f'Registry: {registry}')
  print(f'Prop_dict:\n {prop_dict}')
  print(f'Prop_matrix:\n {prop_matrix}')
  human_choice =np.random.choice(['K','P','N'], p=[1/2,1/4,1/4]) #take_input()

  previous_pos = 0 if previous == 'K' else 1 if previous == 'P' else 2
  curr_pos = 0 if human_choice == 'K' else 1 if human_choice == 'P' else 2

  manage_registry(registry, comp_choice, human_choice)

  prop_dict[previous][curr_pos] += 1

  mianownik = int(prop_dict[previous].sum())
  prop_matrix[previous_pos] = prop_dict[previous]/mianownik

  previous = human_choice
  print('=====')
  print(f'Registry: {registry}')
  print(f'Prop_dict:\n {prop_dict}')
  print(f'Prop_matrix:\n {prop_matrix}')
  comp_choice = np.random.choice(['K','P','N'], p=prop_matrix[curr_pos])


print(peek(registry))  
```
